```python
from skyfield.api import load, Topos
import matplotlib.pyplot as plt
import numpy as np

# 
birth_date = "1998-11-21"
birth_time = "20:20"
birth_location = Topos('29.03 N', '114.54 E')  # 
birth_datetime = f"{birth_date}T{birth_time}Z"

# 
planets = load('de421.bsp')
earth = planets['earth']
sun = planets['sun']
moon = planets['moon']
mercury = planets['mercury']
venus = planets['venus']
mars = planets['mars']
jupiter = planets['jupiter barycenter']
saturn = planets['saturn barycenter']
uranus = planets['uranus barycenter']
neptune = planets['neptune barycenter']
pluto = planets['pluto barycenter']

# 
ts = load.timescale()
t = ts.utc(int(birth_date.split('-')[0]), int(birth_date.split('-')[1]), int(birth_date.split('-')[2]),
           int(birth_time.split(':')[0]), int(birth_time.split(':')[1]))

# 
planet_positions = {
    'Sun': sun,
    'Moon': moon,
    'Mercury': mercury,
    'Venus': venus,
    'Mars': mars,
    'Jupiter': jupiter,
    'Saturn': saturn,
    'Uranus': uranus,
    'Neptune': neptune,
    'Pluto': pluto
}

# 
positions = {}
for name, planet in planet_positions.items():
    astrometric = earth.at(t).observe(planet).apparent()
    ra, dec, distance = astrometric.radec()
    positions[name] = (ra.hours, dec.degrees)

# 
def parse_positions(positions):
    signs = ['Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo', 
             'Libra', 'Scorpio', 'Sagittarius', 'Capricorn', 'Aquarius', 'Pisces']
    sign_degrees = 30
    parsed_positions = {}
    
    for planet, (ra, dec) in positions.items():
        sign_index = int((ra * 15) // sign_degrees) % 12
        degree_in_sign = (ra * 15) % sign_degrees
        parsed_positions[planet] = (signs[sign_index], degree_in_sign, dec)
        
    return parsed_positions

# 
def get_house_info():
    houses = ['1st House', '2nd House', '3rd House', '4th House', '5th House', 
              '6th House', '7th House', '8th House', '9th House', '10th House', 
              '11th House', '12th House']
    house_degrees = 360 / 12
    house_positions = {}
    
    for i, house in enumerate(houses):
        start_degree = i * house_degrees
        end_degree = (i + 1) * house_degrees
        house_positions[house] = (start_degree, end_degree)
    
    return house_positions

# 
parsed_positions = parse_positions(positions)
house_info = get_house_info()

# 
for planet, (sign, degree, dec) in parsed_positions.items():
    for house, (start, end) in house_info.items():
        if start <= (degree % 360) < end:
            current_house = house
            break
    print(f"{planet} is in {sign} at {degree:.2f} degrees with declination {dec:.2f} degrees, in the {current_house}.")

# 
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.set_theta_direction(-1)
ax.set_theta_zero_location('E')

for planet, (ra, dec) in positions.items():
    ax.scatter(np.radians(ra * 15), 90 - dec, label=planet)

# 
for house, (start, end) in house_info.items():
    ax.plot([np.radians(start), np.radians(start)], [0, 90], 'k--', lw=0.5)
    ax.text(np.radians(start + 15), 85, house, horizontalalignment='center', size=8, color='gray')

ax.legend()
ax.set_title('Astrological Chart')
plt.show()
```
